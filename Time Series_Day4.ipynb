```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

# Generate a synthetic non-stationary time series
np.random.seed(42)
time = pd.date_range(start='2022-01-01', end='2023-12-31', freq='D')
data = np.cumsum(np.random.normal(size=len(time)))

# Create a DataFrame
df = pd.DataFrame({'date': time, 'value': data})
df.set_index('date', inplace=True)

# Plot the time series
plt.plot(df['value'])
plt.title('Non-Stationary Time Series')
plt.show()

# Perform ADF test
result = adfuller(df['value'])

# Extract results
adf_statistic = result[0]
p_value = result[1]
critical_values = result[4]

# Display results
print(f'ADF Statistic: {adf_statistic}')
print(f'P-value: {p_value}')
print(f'Critical Values: {critical_values}')

# Interpret the results
if p_value < 0.05:
    print('Reject the null hypothesis. The time series is likely stationary.')
else:
    print('Fail to reject the null hypothesis. The time series is likely non-stationary.')
```

```python
# Generate a synthetic stationary time series
np.random.seed(42)
time = pd.date_range(start='2022-01-01', end='2023-12-31', freq='D')
data = np.random.normal(size=len(time))

# Create a DataFrame
df = pd.DataFrame({'date': time, 'value': data})
df.set_index('date', inplace=True)

# Plot the time series
plt.plot(df['value'])
plt.title('Stationary Time Series')
plt.show()

# Perform ADF test
result = adfuller(df['value'])

# Extract results
adf_statistic = result[0]
p_value = result[1]
critical_values = result[4]

# Display results
print(f'ADF Statistic: {adf_statistic}')
print(f'P-value: {p_value}')
print(f'Critical Values: {critical_values}')

# Interpret the results
if p_value < 0.05:
    print('Reject the null hypothesis. The time series is likely stationary.')
else:
    print('Fail to reject the null hypothesis. The time series is likely non-stationary.')
```

### Decomposing the time series data

```python
#Data clearly has a linear uptrend.
#A clear cut seasonal pattern of a 1 year duration is present in the data.
from statsmodels.tsa.seasonal import seasonal_decompose

dec = seasonal_decompose(idd,period = 12,model='multiplicative').plot()
plt.show()
```

```python
#Extracting the values for trend, seasonal and residual from decomposed data
decomposed = seasonal_decompose(idd,period = 12,model='multiplicative')
orginal_vs_decomposed_series = pd.concat([decomposed.observed, decomposed.trend,decomposed.seasonal,decomposed.resid], axis=1)
orginal_vs_decomposed_series.columns = ['Actual','Trend','Seasonal','Residual Or Unexplained']
orginal_vs_decomposed_series.to_csv('orginal_vs_decomposed_series.csv')
```

```python
orginal_vs_decomposed_series = orginal_vs_decomposed_series.reset_index()
orginal_vs_decomposed_series.head()
```

```python
#Adding rank column in decomposed series
# creating a rank column and passing the returned rank series
orginal_vs_decomposed_series["Rank"] = orginal_vs_decomposed_series["Month"].rank()
```

```python
orginal_vs_decomposed_series.head()
```

```python
# split into X and y
data_for_model = orginal_vs_decomposed_series.dropna()
X = data_for_model.loc[:, ['Rank']]
Y  = data_for_model['Trend']

# fitting a linear model
import statsmodels.api as sm
X = sm.add_constant(X)
model = sm.OLS(Y,X).fit()
```

```python
print(model.summary())
```

```python
#Predicied Trend Values
orginal_vs_decomposed_series['Predicted_trend']=(orginal_vs_decomposed_series['Rank']*2.6669 + 84.6483)
orginal_vs_decomposed_series.head()
```

```python
orginal_vs_decomposed_series['Predicted_timeseries']=orginal_vs_decomposed_series['Predicted_trend']*
orginal_vs_decomposed_series['Seasonal']
```

```python
plt.plot(orginal_vs_decomposed_series['Actual'])
plt.plot(orginal_vs_decomposed_series['Predicted_timeseries'], color='red')
print('Plotting Basic Model')
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
error = np.sqrt(mean_squared_error(orginal_vs_decomposed_series['Actual'],
                                   orginal_vs_decomposed_series['Predicted_timeseries']))
print('Total RMSE: %.2f' % error)
```

```python
# Split the data into training and testing sets
train = data[:-10]  # Use all but the last 10 observations for training
test = data[-10:]  # Use the last 10 observations for testing
```

```python
#Standard ARIMA Model
!pip install pmdarima --quiet
import pmdarima as pm
SARIMA_model = pm.auto_arima(train['#Passengers'],
                      start_p=0,
                      d=1,
                      start_q=0,
                      m=12, # frequency of series (if m==1, seasonal is set to FALSE automatically)
                      seasonal=True)

SARIMA_model.order
#this shows that p value is 1, d value is 1 and q value is 0
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Make predictions on the testing set
predictions = SARIMA_model.predict(n_periods=len(test))

# Calculate evaluation metrics
mse = mean_squared_error(test['#Passengers'], predictions)
mae = mean_absolute_error(test['#Passengers'], predictions)

print(f"Root Mean Squared Error (RMSE): {np.sqrt(mse)}")
print(f"Mean Absolute Error (MAE): {mae}")
```

```python
def forecast(ARIMA_model, periods=24):
    # Forecast
    n_periods = periods
    fitted, confint = ARIMA_model.predict(n_periods=n_periods, return_conf_int=True)
    index_of_fc = pd.date_range(data.index[-1] + pd.DateOffset(months=1), periods = n_periods, freq='MS')

    # make series for plotting purpose
    fitted_series = pd.Series(fitted, index=index_of_fc)
    # Plot
    plt.figure(figsize=(15,7))
    plt.plot(data["#Passengers"], color='#1f76b4')
    plt.plot(fitted_series, color='darkgreen')
    plt.title("ARIMA/SARIMA - Forecast of Airline Passengers")
    plt.show()

forecast(SARIMA_model)

```
